```python
import matplotlib.pyplot as plt

x=['Mon','Tue','Wed','Thur','Fri','Sat']

y=[10,15,12,7,5,25]

plt.plot(x,y)
plt.title("Bakery sales this week")

plt.xlabel("Days of week")
plt.ylabel("Sales per Day")

plt.show()
```

```python
import matplotlib.pyplot as plt

x=['Mon','Tue','Wed','Thur','Fri','Sat']

y=[10,15,12,7,5,25]

plt.plot(x,y,color="red",linestyle='--',linewidth=2,marker='o',label='2025 sales')
plt.xlim(0,5)
plt.ylim(5,30)
plt.show()
```
